```python
import numpy as np
import pandas as pd 
import seaborn as sns
```

```python
#Uploading datasets
testSet = pd.read_csv("test.csv")
testSet.set_index('PassengerId',inplace=True)
trainingSet = pd.read_csv("train.csv")
trainingSet.set_index('PassengerId',inplace=True)

trainingSet.head()
```

```python
#Exploratory Data Analysis

#Univariate statistics
nb_passagers = 891
survivors = trainingSet[trainingSet['Survived'] == 1]['Survived'].count()
deaths = trainingSet[trainingSet['Survived'] == 0]['Survived'].count()


print("Number of passengers registered : "+str(nb_passagers))
print("Number of survivors : "+str(survivors))
print("Number of deaths : "+str(deaths))
print("Survival rate : "+str(survivors*100/(survivors+deaths))+"%")

sns.countplot(x=trainingSet["Survived"])
```

```python
firstClass = trainingSet[trainingSet['Pclass'] == 1]['Pclass'].count()
secondClass = trainingSet[trainingSet['Pclass'] == 2]['Pclass'].count()
thirdClass = trainingSet[trainingSet['Pclass'] == 3]['Pclass'].count()
classTotal = trainingSet['Pclass'].count()

print("Number of first class passengers : "+str(firstClass)+ " (" + str(firstClass*100/classTotal) + "% of the passengers)")
print("Number of second class passengers : "+str(secondClass)+ " (" + str(secondClass*100/classTotal) + "% of the passengers)")
print("Number of third class passengers : "+str(thirdClass)+ " (" + str(thirdClass*100/classTotal) + "% of the passengers)")

sns.countplot(x=trainingSet["Pclass"])
```

```python
male = trainingSet[trainingSet['Sex'] == 'male']['Sex'].count()
female = trainingSet[trainingSet['Sex'] == 'female']['Sex'].count()

print("Number of male passengers : "+str(male)+ " (" + str(male*100/nb_passagers) + "% of the passengers)")
print("Number of female passengers : "+str(female)+ " (" + str(female*100/nb_passagers) + "% of the passengers)")

sns.countplot(x=trainingSet["Sex"])
```

```python
Cherbourg = trainingSet[trainingSet['Embarked'] == "C"]['Embarked'].count()
Queenstown = trainingSet[trainingSet['Embarked'] == "Q"]['Embarked'].count()
Southampton = trainingSet[trainingSet['Embarked'] == "S"]['Embarked'].count()


print("Number of passagers who embarked from Cherbourg : "+str(Cherbourg)+ " (" + str(Cherbourg*100/nb_passagers) + "% of the passengers)")
print("Number of passagers who embarked from Queenstown : "+str(Queenstown)+ " (" + str(Queenstown*100/nb_passagers) + "% of the passengers)")
print("Number of passagers who embarked from Southampton : "+str(Southampton)+ " (" + str(Southampton*100/nb_passagers) + "% of the passengers)")

sns.countplot(x=trainingSet["Embarked"])
```

```python
#Coding quantitative variables
#sex : 0 if male, 1 if female
trainingSet.loc[trainingSet['Sex']=='male','Sex'] = 0
trainingSet.loc[trainingSet['Sex']=='female','Sex'] = 1

#Classes, creating a boolean column for each
trainingSet['First Class'] = trainingSet['Pclass'].apply(lambda x: 1 if x==1 else 0)
trainingSet['Second Class'] = trainingSet['Pclass'].apply(lambda x: 1 if x==2 else 0)
trainingSet['Third Class'] = trainingSet['Pclass'].apply(lambda x: 1 if x==3 else 0)

#Age
trainingSet['Young (0-20 years old)'] = trainingSet['Age'].apply(lambda x: 1 if x<=20.125 else 0)
trainingSet['Young Adult (20-28 years old)'] = trainingSet['Age'].apply(lambda x: 1 if x>20.125 and x<=28 else 0)
trainingSet['Adult (28-38 years old)'] = trainingSet['Age'].apply(lambda x: 1 if x>28 and x<=38 else 0)
trainingSet['Old (38+ years old)'] = trainingSet['Age'].apply(lambda x: 1 if x>38  else 0)

#Siblings
trainingSet['No Siblings'] = trainingSet['SibSp'].apply(lambda x: 1 if x==0 else 0)
trainingSet['Have Siblings'] = trainingSet['SibSp'].apply(lambda x: 1 if x>0 else 0)

#Parents and chikdren
trainingSet['No Parents or Children'] = trainingSet['Parch'].apply(lambda x: 1 if x==0 else 0)
trainingSet['Have Parents or Children'] = trainingSet['Parch'].apply(lambda x: 1 if x>0 else 0)

#Ticket Fare

trainingSet['Very Cheap (<7.9$)'] = trainingSet['Fare'].apply(lambda x: 1 if x<=7.9104 else 0)
trainingSet['Cheap (7.9-14.45$)'] = trainingSet['Age'].apply(lambda x: 1 if x>7.9104 and x<=14.4542 else 0)
trainingSet['Normal (14.45-31$)'] = trainingSet['Age'].apply(lambda x: 1 if x>14.4542 and x<=31 else 0)
trainingSet['Expensive (>31$)'] = trainingSet['Age'].apply(lambda x: 1 if x>31  else 0)

trainingSet.head()
```

```python
#Multivariate analysis

for column in trainingSet.columns:
    if column != "PassengerId" and column != "Survived" and column != "Name" and column != "Age" and column != "Fare" and column != "SibSp" and column != "Parch" and column != "Ticket" and column != "Cabin":
        print(pd.crosstab(trainingSet["Survived"],trainingSet[column]))
```
